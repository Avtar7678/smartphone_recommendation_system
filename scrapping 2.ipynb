```python
import pandas as pd
import requests
import bs4
from bs4 import BeautifulSoup as bs
```

```python
phonename=[]
phoneprice=[]
phonefeature=[]

pagenumber=input('Enter the page number')
for i in range(1,int(pagenumber)+1):
    url="https://www.flipkart.com/search?q=smartphone+under+20000&as=on&as-show=on&otracker=AS_Query_HistoryAutoSuggest_1_17_na_na_na&otracker1=AS_Query_HistoryAutoSuggest_1_17_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=smartphone+under+20000&requestId=f8e8f2ed-a083-4b0d-b842-37a8f7f9c56e&as-searchtext=smartphone+under+%3D2&page="+str(i)
    req=requests.get(url)
    content=bs(req.content,'html.parser')
    name=(content.find_all('div',class_="_4rR01T"))
    price=(content.find_all('div',class_="_3tbKJL"))
    feature=(content.find_all('ul',class_="_1xgFaf"))
    
    print('Phone in page' +str(i))
    print(len(name))
    print('Phone in page '+str(i))
    print(len(price))
    print('phone in page'+str(i))
    print(len(feature))
    
    
    
    
    
    for i in name:
        phonename.append(i.text)
        print(phonename)
    for i in price:
        phoneprice.append(i.text)
        print(phoneprice)    
    for i in feature:
        phonefeature.append(i.text)
        print(phonefeature)
    
    
    
```

```python
for i in phonename:
    print(i)
for i in phoneprice:
    print(i)
for i in phonefeature:
    print(i)
```

```python

import pandas as pd
data=({"phonename":phonename,"phoneprice":phoneprice,"phonefeature":phonefeature})
df=pd.DataFrame(data)
print(df)
```
